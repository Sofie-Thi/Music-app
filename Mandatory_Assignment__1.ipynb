```python
from sklearn.linear_model import LinearRegression
import numpy as np

lin_reg = LinearRegression()

X = wineData['chlorides'].values
y = wineData['quality'].values

learning_rate = 1.0
n_iterations = 5000
theta = np.zeros(2)

###

# Add bias term (column of ones)
X_b = np.column_stack([np.ones(len(X)), X])

# Gradient descent parameters
learning_rate = 1.00
n_iterations = 10000
theta = np.zeros(2)  # [intercept, slope]

# Gradient descent
for i in range(n_iterations):
    predictions = X_b.dot(theta)
    errors = predictions - y
    gradient = X_b.T.dot(errors) / len(y)
    theta -= learning_rate * gradient

    # Print progress every 100 iterations
    if i % 100 == 0:
        cost = np.mean(errors**2)
        #print(f"Iteration {i}: Cost = {cost:.4f}")

# Final parameters
intercept, slope = theta
print(f"\nFinal equation: quality = {intercept:.2f} + {slope:.2f} * chlorides")

# Make predictions
predictions = X_b.dot(theta)

# Calculate R-squared
ss_res = np.sum((y - predictions) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)
r_squared = 1 - (ss_res / ss_tot)
print(f"R-squared: {r_squared:.3f}")

# Simple visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6, label='Data points')

# Plot regression line
x_line = np.linspace(X.min(), X.max(), 100)
y_line = intercept + slope * x_line
plt.plot(x_line, y_line, 'r-', linewidth=3, label='Regression line')

plt.xlabel('Chlorides')
plt.ylabel('Quality')
plt.title('Wine Quality og chlorides')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import KFold

#størslse på testgruppene (...)
kf = KFold(n_splits=5)
for train, test in kf.split(train_set):
    print('testgruppe lengde',len(train))
    #print("%s %s" % (train, test))
```

```python
# Use the linear regression function from earlier
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None
        self.cost_history = []

    #Train the model using gradient descent
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.theta = np.zeros((n_features, 1))

        for i in range(self.n_iterations):
            predictions = X @ self.theta
            error = predictions - y.reshape(-1, 1)
            gradient = (1/n_samples) * X.T @ error
            self.theta -= self.learning_rate * gradient

            # Store cost
            cost = (1/(2*n_samples)) * np.sum(error**2)
            self.cost_history.append(cost)

        return self
    #Make predictions using the trained model
    def predict(self, X):
        return X @ self.theta

    #Calculate R-squared score
    def score(self, X, y):
        predictions = self.predict(X)
        ss_res = np.sum((y.reshape(-1, 1) - predictions) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)

# Initialize KFold for 5 folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store results for both features
results_alcohol = []
results_chlorides = []

print("5-FOLD CROSS VALIDATION RESULTS")

for fold, (train_idx, test_idx) in enumerate(kf.split(wineData), 1):
    print(f"\n--- Fold {fold} ---")

    # Split data
    X_train_alc = wineData['alcohol'].values[train_idx].reshape(-1, 1)
    X_test_alc = wineData['alcohol'].values[test_idx].reshape(-1, 1)
    X_train_chl = wineData['chlorides'].values[train_idx].reshape(-1, 1)
    X_test_chl = wineData['chlorides'].values[test_idx].reshape(-1, 1)
    y_train = wineData['quality'].values[train_idx]
    y_test = wineData['quality'].values[test_idx]

    # Add bias term
    X_train_alc_b = np.column_stack([np.ones(len(X_train_alc)), X_train_alc])
    X_test_alc_b = np.column_stack([np.ones(len(X_test_alc)), X_test_alc])
    X_train_chl_b = np.column_stack([np.ones(len(X_train_chl)), X_train_chl])
    X_test_chl_b = np.column_stack([np.ones(len(X_test_chl)), X_test_chl])

    # Train alcohol model
    model_alc = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
    model_alc.fit(X_train_alc_b, y_train)

    # Train chlorides model
    model_chl = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
    model_chl.fit(X_train_chl_b, y_train)

    # Make predictions
    y_pred_alc_train = model_alc.predict(X_train_alc_b)
    y_pred_alc_test = model_alc.predict(X_test_alc_b)
    y_pred_chl_train = model_chl.predict(X_train_chl_b)
    y_pred_chl_test = model_chl.predict(X_test_chl_b)

    # Calculate metrics function
    def calculate_metrics(y_true, y_pred):
        mse = np.mean((y_true - y_pred.flatten()) ** 2)
        rmse = np.sqrt(mse)
        r2 = 1 - (np.sum((y_true - y_pred.flatten()) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2))
        return {'mse': mse, 'rmse': rmse, 'r2': r2}

    # Calculate metrics
    alc_train_metrics = calculate_metrics(y_train, y_pred_alc_train)
    alc_test_metrics = calculate_metrics(y_test, y_pred_alc_test)
    chl_train_metrics = calculate_metrics(y_train, y_pred_chl_train)
    chl_test_metrics = calculate_metrics(y_test, y_pred_chl_test)

    # Store results
    results_alcohol.append({
        'fold': fold,
        'train_metrics': alc_train_metrics,
        'test_metrics': alc_test_metrics,
        'theta': model_alc.theta.flatten()
    })

    results_chlorides.append({
        'fold': fold,
        'train_metrics': chl_train_metrics,
        'test_metrics': chl_test_metrics,
        'theta': model_chl.theta.flatten()
    })

    print(f"Alcohol - Test R²: {alc_test_metrics['r2']:.3f}, RMSE: {alc_test_metrics['rmse']:.3f},MSE: {alc_test_metrics['mse']:.3f}v")
    print(f"Chlorides - Test R²: {chl_test_metrics['r2']:.3f}, RMSE: {chl_test_metrics['rmse']:.3f}, MSE: {chl_test_metrics['mse']:.3f}")

# Calculate mean and variance
def calculate_statistics(results, feature_name):
    test_r2 = [r['test_metrics']['r2'] for r in results]
    test_rmse = [r['test_metrics']['rmse'] for r in results]
    test_mse = [r['test_metrics']['mse'] for r in results]

    return {
        'feature': feature_name,
        'mean_r2': np.mean(test_r2),
        'var_r2': np.var(test_r2),
        'mean_rmse': np.mean(test_rmse),
        'var_rmse': np.var(test_rmse),
        'mean_mse': np.mean(test_mse),
        'var_mse': np.var(test_mse)
    }

stats_alc = calculate_statistics(results_alcohol, 'alcohol')
stats_chl = calculate_statistics(results_chlorides, 'chlorides')

```

1.5 Multiple Linear Regression


```python
from sklearn.preprocessing import StandardScaler

df = wineData

# Define features and target
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
           'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
           'pH', 'sulphates', 'alcohol']
target = 'quality'

X = df[features].values
y = df[target].values

# Extended Linear Regression class with multiple features support
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, n_iterations=1000, fit_intercept=True):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.fit_intercept = fit_intercept
        self.theta = None
        self.cost_history = []
        self.theta_history = []

    def fit(self, X, y, verbose=False):
        if self.fit_intercept:
            X = np.column_stack([np.ones(X.shape[0]), X])

        n_samples, n_features = X.shape
        self.theta = np.zeros(n_features)
        self.theta_history.append(self.theta.copy())

        for i in range(self.n_iterations):
            predictions = X @ self.theta
            error = predictions - y
            gradient = (1/n_samples) * X.T @ error
            self.theta -= self.learning_rate * gradient

            # Store history
            cost = (1/(2*n_samples)) * np.sum(error**2)
            self.cost_history.append(cost)
            self.theta_history.append(self.theta.copy())

            if verbose and i % 200 == 0:
                print(f"Iteration {i}: Cost = {cost:.4f}")

        return self


    def predict(self, X):
        if self.fit_intercept:
            X = np.column_stack([np.ones(X.shape[0]), X])
        return X @ self.theta

    def score(self, X, y):
        predictions = self.predict(X)
        ss_res = np.sum((y - predictions) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)

# Initialize KFold for 5 folds (same splits as before)
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store results
results_simple_alc = []
results_simple_chl = []
results_multiple = []

for fold, (train_idx, test_idx) in enumerate(kf.split(df), 1):
    print(f"\n--- Fold {fold} ---")

    # Split data
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Scale features for multiple regression
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # For simple regression (using same splits)
    X_train_alc = df['alcohol'].values[train_idx].reshape(-1, 1)
    X_test_alc = df['alcohol'].values[test_idx].reshape(-1, 1)
    X_train_chl = df['chlorides'].values[train_idx].reshape(-1, 1)
    X_test_chl = df['chlorides'].values[test_idx].reshape(-1, 1)

    # Calculate metrics function
    def calculate_metrics(y_true, y_pred):
        mse = np.mean((y_true - y_pred) ** 2)
        rmse = np.sqrt(mse)
        r2 = 1 - (np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2))
        return {'mse': mse, 'rmse': rmse, 'r2': r2}

    # 1. Simple Linear Regression - Alcohol
    model_simple_alc = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
    model_simple_alc.fit(X_train_alc, y_train)
    y_pred_simple_alc = model_simple_alc.predict(X_test_alc)
    metrics_simple_alc = calculate_metrics(y_test, y_pred_simple_alc)
    results_simple_alc.append(metrics_simple_alc)

    # 2. Simple Linear Regression - Chlorides
    model_simple_chl = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
    model_simple_chl.fit(X_train_chl, y_train)
    y_pred_simple_chl = model_simple_chl.predict(X_test_chl)
    metrics_simple_chl = calculate_metrics(y_test, y_pred_simple_chl)
    results_simple_chl.append(metrics_simple_chl)

    # 3. Multiple Linear Regression - All Features
    model_multiple = LinearRegressionGD(learning_rate=0.1, n_iterations=2000)
    model_multiple.fit(X_train_scaled, y_train)
    y_pred_multiple = model_multiple.predict(X_test_scaled)
    metrics_multiple = calculate_metrics(y_test, y_pred_multiple)
    results_multiple.append(metrics_multiple)

    print(f"Simple (Alcohol)  - R²: {metrics_simple_alc['r2']:.3f}, RMSE: {metrics_simple_alc['rmse']:.3f}")
    print(f"Simple (Chlorides)- R²: {metrics_simple_chl['r2']:.3f}, RMSE: {metrics_simple_chl['rmse']:.3f}")
    print(f"Multiple (All)     - R²: {metrics_multiple['r2']:.3f}, RMSE: {metrics_multiple['rmse']:.3f}")

# Calculate statistics
def calculate_final_stats(results, model_name):
    r2_scores = [r['r2'] for r in results]
    rmse_scores = [r['rmse'] for r in results]
    mse_scores = [r['mse'] for r in results]

    return {
        'model': model_name,
        'mean_r2': np.mean(r2_scores),
        'std_r2': np.std(r2_scores),
        'mean_rmse': np.mean(rmse_scores),
        'std_rmse': np.std(rmse_scores),
        'mean_mse': np.mean(mse_scores),
        'std_mse': np.std(mse_scores)
    }

stats_simple_alc = calculate_final_stats(results_simple_alc, 'Simple (Alcohol)')
stats_simple_chl = calculate_final_stats(results_simple_chl, 'Simple (Chlorides)')
stats_multiple = calculate_final_stats(results_multiple, 'Multiple (All Features)')

```
